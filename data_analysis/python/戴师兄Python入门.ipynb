````python
%matplotlib notebook
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']#
plt.rcParams['axes.unicode_minus'] = False #
````

````python
import pandas as pd
olpc = pd.read_csv('Tokyo 2021 dataset.csv') #CSV
olpc.info()
````

````python
top5 = olpc[olpc.Rank<=5] #>5top5
top5.rename(columns={'Team/NOC':'Team'},inplace=True) #Team/NOC  Team
top5
````

###   
figure()

````python
plt.figure()
plt.plot(top5['NOCCode'],top5['Gold Medal'])#plotx,y
````

````python
plt.figure()
plt.bar(top5['NOCCode'],top5['Gold Medal'])#barx,y
````

````python
plt.figure()
plt.barh(top5['NOCCode'],top5['Gold Medal']) #barhx,y
````

````python
plt.figure()
plt.pie(top5['Gold Medal']) #pie Seires
````

````python
plt.figure()
plt.scatter(top5['NOCCode'],top5['Gold Medal']) #scatter 
````

````python
plt.figure()
plt.hist(top5['Gold Medal'],bins=5) #histSeriresbins
````

### 

````python
plt.figure()
plt.plot(top5['NOCCode'],top5['Gold Medal'])
plt.ylim(20,40) #y
#x plt.xlim()
````

````python
plt.figure()
plt.plot(top5['NOCCode'],top5['Gold Medal'])
plt.xlabel('')
plt.ylabel('')
#
````

### 

````python
plt.figure()
plt.plot(top5['NOCCode'],top5['Gold Medal'],label = '') # label
plt.legend() #
#
````

### 

````python
plt.figure()
plt.plot(top5['NOCCode'],top5['Gold Medal'])
for x,y in zip(top5['NOCCode'],top5['Gold Medal']):
    plt.text(x,y+0.1,'%d' % y,ha = 'center',va = 'bottom')
#
````
